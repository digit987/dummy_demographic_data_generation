```python
def get_random_area_sector_activity():
    random_work_area_generator = random_sector_generator = random_activity_generator = ''
    random_work_area_generator = random.choice(work_area)
    random_work_area_index = work_area.index(random_work_area_generator)
    if len(sector[random_work_area_index]) == 0:
        random_sector_generator = ''
        random_activity_generator = ''
    else:
        random_sector_generator = random.choice(sector[random_work_area_index])
        random_sector_index = sector[random_work_area_index].index(random_sector_generator)
        if len(activity[random_work_area_index][random_sector_index]) == 0:
            random_activity_generator = ''
        else:
            random_activity_generator = random.choice(activity[random_work_area_index][random_sector_index])
    return random_work_area_generator, random_sector_generator, random_activity_generator
```

```python
def random_aadhar_generator():
    random_aadhar = ''
    for i in range(12):
        if i == 0:
            random_aadhar += str(random.randrange(1,9))
        else:
            random_aadhar += str(random.randrange(0,9))
    return random_aadhar
```
